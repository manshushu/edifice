```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QGridLayout
from PyQt5.QtCore import Qt

app = QApplication(sys.argv)

window = QWidget()
window.setWindowTitle('人体模型示意图')

grid_layout = QGridLayout()

head_label = QLabel('头')
head_label.setToolTip('这里是头')
grid_layout.addWidget(head_label, 0, 1)
neck_label = QLabel('颈')
neck_label.setToolTip('这里是颈')
grid_layout.addWidget(neck_label, 1, 1)
r_arm_label = QLabel('右臂')
r_arm_label.setToolTip('这里是右臂')
r_arm_label.setProperty('part', 'right arm')  # 设置自定义属性
grid_layout.addWidget(r_arm_label, 2, 0)
chest_label = QLabel('胸')
chest_label.setToolTip('这里是胸')
grid_layout.addWidget(chest_label, 2, 1)
l_arm_label = QLabel('左臂')
l_arm_label.setToolTip('这里是左臂')
l_arm_label.setProperty('part', 'left arm')  # 设置自定义属性
grid_layout.addWidget(l_arm_label, 2, 2)
waist_label = QLabel('腰')
waist_label.setToolTip('这里是腰')
grid_layout.addWidget(waist_label, 3, 1)
r_leg_label = QLabel('右腿')
r_leg_label.setToolTip('这里是右腿')
r_leg_label.setProperty('part', 'right leg')  # 设置自定义属性
grid_layout.addWidget(r_leg_label, 4, 0)
l_leg_label = QLabel('左腿')
l_leg_label.setToolTip('这里是左腿')
l_leg_label.setProperty('part', 'left leg')  # 设置自定义属性
grid_layout.addWidget(l_leg_label, 4, 2)

def on_part_clicked():
    # 获取当前部位的标签对象以及其自定义属性
    label = window.sender()
    part = label.property('part')
    # 弹出相应提示信息
    print(f'{part} 被点击了')

head_label.mousePressEvent = on_part_clicked
neck_label.mousePressEvent = on_part_clicked
r_arm_label.mousePressEvent = on_part_clicked
chest_label.mousePressEvent = on_part_clicked
l_arm_label.mousePressEvent = on_part_clicked
waist_label.mousePressEvent = on_part_clicked
r_leg_label.mousePressEvent = on_part_clicked
l_leg_label.mousePressEvent = on_part_clicked

window.setLayout(grid_layout)

window.show()
sys.exit(app.exec_())
```
