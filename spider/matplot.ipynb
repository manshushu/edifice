```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei']
```

```python
df = pd.read_csv('d:\\electricityConsumptionAndProductioction.csv')
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
sns.histplot(data=df, x='Consumption')
plt.grid(linestyle='-',linewidth='1')
plt.show()
```

```python
sns.kdeplot(data=df, x='Consumption')
plt.grid(linestyle='-',linewidth='1')
plt.legend(['count'])
plt.show()
```

```python
sns.boxplot(data=df, x='Consumption')
plt.grid(linestyle='-',linewidth='1')
plt.show()
```

```python
print(df.dtypes)
```

```python
import matplotlib.ticker as ticker
df.plot(x='DateTime', y='Consumption')
plt.title("Consumption时序图")
plt.xlabel("time")
plt.ylabel("Consumption")
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(12))
ticker.MAXTICKS =10000
plt.show()
```
