```python
import cv2
import numpy as np
from keras.models import load_model

# 加载保存好的模型
model = load_model('cats_dogs.h5')

# 捕获默认摄像头的视频流
cap = cv2.VideoCapture(0)

while True:
    # 从视频流中读取一帧图像
    ret, frame = cap.read()
    if ret:
        x, y, w, h = 200, 200, 150, 150   # ROI左上角坐标：(x,y)，宽度w，高度h
        cv2.rectangle(frame, (x,y), (x+w,y+h), (255,0,0), 3)
        cv2.line(frame, (x,y), (x+w,y+h), (255,0,0), 3)
        cv2.line(frame, (x+w,y), (x,y+h), (255,0,0), 3)
    
    # 获取ROI区域并进行对应的处理
        roi = frame[y:y+h, x:x+w]
    # 对输入图像进行预处理
        img = cv2.resize(roi, (150, 150))
        img = np.expand_dims(img, axis=0) / 255.0
    
    # 使用模型进行预测
        prediction = model.predict(img)
        predictions_list = prediction.tolist()

    # 根据预测值进行相应操作
        if predictions_list[0][0]<predictions_list[0][1]:
            # 如果模型预测为负类别(dog)，做出对应操作
            cv2.putText(frame, 'This is a dog', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
        else:
            # 如果模型预测为正类别(cat)，做出对应操作
            cv2.putText(frame, 'This is a cat', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # 显示预测结果
        cv2.imshow('Prediction', frame)
    
    # 退出循环条件
    if cv2.waitKey(1) == ord('q'):
        break
        
# 关闭视频流和窗口
cap.release()
cv2.destroyAllWindows()
```
